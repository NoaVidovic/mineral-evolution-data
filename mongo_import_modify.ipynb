```python
import pymongo
import pandas as pd
```

```python
client = pymongo.MongoClient('localhost', 27017)
db = client.mineral_db
```

```python
pgm = pd.read_csv('2021_hazen_pgm.csv',
                  index_col='Mineral Name',
                  dtype=str,
                 ).transpose().fillna(0).replace('1', 1)
pgm
```

```python
# replace with pgm[mineral][mode] when the table is fixed
def p(mineral, mode):
    try:
        return pgm[mineral][mode]
    except:
        return None
```

```python
minerals = pgm.columns[1:]
modes = pgm.transpose().columns[:-3]
```

```python
minerals_in_pgm = { mode :
                   [ mineral for mineral in pgm.columns[1:] if p(mineral, mode) ]
                   for mode in pgm.transpose().columns[:-3]
                  }
minerals_in_pgm
```

```python
mineral_pgms = { mineral:
                [ mode for mode in modes if p(mineral, mode) == 1 ]
                for mineral in minerals
               }
mineral_pgms
```
